```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

### import Dataset

```python
df = pd.read_csv("Dataset.csv")
```

```python
df.info()
```

### check whether duplicates are present

```python
df.duplicated()
```

### check null values

```python
df.isnull().any()
```

### Replacing null values with unknown

```python
df['Postal Code'] = df['Postal Code'].fillna('Unknown')
```

```python
df['Ship Date'].dtype
```

### Converting `Order Date `and `ship Date` into datetime format

```python
df['Order Date'] =  pd.to_datetime(df['Order Date'],format = 'mixed',dayfirst=True, errors ='coerce')
df['Ship Date'] =  pd.to_datetime(df['Ship Date'],format = 'mixed',dayfirst=True, errors ='coerce')
```

### Re-check  whether any unchanged or null values

```python
df['Order Date'].isnull().sum()
df['Ship Date'].isnull().sum()
```

### Adding year and month column

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

### Adding Delivery Days column

```python
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

### Analysing 

```python
df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
df.groupby('Segment')['Sales'].sum()
```

### Converting column names to snake_case

```python
df.columns = df.columns.str.replace(' ','_')
```
